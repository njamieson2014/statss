```python
# Define the parameters for the normal distribution
mean = 0  # Mean (mu) of the distribution
std_dev = 0.1  # Standard deviation (sigma) of the distribution

# Generate 1000 data points following a normal distribution with the specified mean and standard deviation
sample_size = 10000
data_points = np.random.normal(mean, std_dev, sample_size)

# Plot the histogram of the data points
bins_number = 30  # Number of bins for the histogram
hist_count, x, ignored = plt.hist(data_points, bins_number, density=True)

# Plot the probability density function of the normal distribution
normal_dist_curve = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(- (x - mean) ** 2 / (2 * std_dev ** 2))
plt.plot(x, normal_dist_curve, linewidth=2, color='r')

# Set the title and labels for the plot
plt.title('Normal Distribution Visualization')
plt.xlabel('Value')
plt.ylabel('Probability Density')

# Display the plot
plt.show()
```

```python
def mean(x):
    return np.sum(x)/x.shape[0]
```

```python
def var(x):
    return mean((x - mean(x))**2)
```

```python
def std(x):
    return np.sqrt(var(x))
```

```python
def median(x): #median of the x values
    n = len(x)
    sorted_x = np.sort(x)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_x[mid - 1] + sorted_x[mid]) / 2
    else:
        return sorted_x[mid]
```

```python
def mode(x): #most common value
    values, counts = np.unique(x, return_counts=True)
    max_count_index = np.argmax(counts)
    return values[max_count_index]
```

```python
def range1(x): #range of values
    return np.max(x) - np.min(x)
```

```python
data = np.array([1, 2, 2, 3, 4, 5, 5, 5, 6])

# Testing the functions
mean_value = mean(data)
var_value = var(data)
std_value = std(data)
median_value = median(data)
mode_value = mode(data)
range_value = range1(data)

mean_value, var_value, std_value, median_value, mode_value, range_value
```

```python
inside_circle  = points[:,0]**2 + points[:,1]**2  <=  1 #checks if the randomly generated point is greater than or less than 1 units away from the origin (vector magnitude)
outside_circle = points[:,0]**2 + points[:,1]**2  > 1

plt.plot(points[inside_circle,0],points[inside_circle,1],'g.') #colors points that have distance from origin less than or equal to one as green
plt.plot(points[outside_circle,0],points[outside_circle,1],'r.') #colors points that have distance from the origin greater than one as red


plt.gca().set_aspect(1)
```

# Complete Code for Estimating  using Monte Carlo Simulation

```python
# Number of random points to generate
num_points = 10000

# Generating random points
x_points = np.random.uniform(-1, 1, num_points)
y_points = np.random.uniform(-1, 1, num_points)

# Calculating the number of points inside the quarter circle
points_inside = np.sqrt(x_points**2 + y_points**2) <= 1
num_inside = np.sum(points_inside)

# Estimating 
pi_estimate = 4 * num_inside / num_points

#same process as above, but now plots the results and the target area is circled in red
# Plotting the points and the quarter circle
plt.figure(figsize=(6, 6))
plt.scatter(x_points[points_inside], y_points[points_inside], color='green', label='Inside')
plt.scatter(x_points[~points_inside], y_points[~points_inside], color='red', label='Outside')
circle = plt.Circle((0, 0), 1, color='blue', fill=False)
plt.gca().add_artist(circle)
plt.title('Estimating  using Monte Carlo Simulation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axis('equal')
plt.show()

pi_estimate
```

# e from Random Numbers

```python
X = np.random.random((5,5)) #generates an array of random numbers
X
```

```python
Y = np.cumsum(X,1) #cumulative sum of each row in array X, 1= rows, 0 = columns
Y
```

```python
Z = np.argmax(Y > 1,1) + 1 #finds in rows in array Y, the index of the first value which is greater than 1, for all rows, then adds 1
Z #if value = 1, the argmax outputted a 0 for false
```

```python
np.mean(Z)
```

```python
np.mean(np.argmax(np.cumsum(np.random.random((10000000,10)),1) > 1,1) + 1) #combine it all together, the more N terms, the closer to e
```
